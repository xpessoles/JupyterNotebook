### Préliminaires

#### Question 1 : Donner dans le domaine temporel une relation liant $e_v(t)$, $\varepsilon_v(t)$ et $s_v(t)$.

#### Réponse 1
La relation du comparateur nous donne : $\varepsilon_v(t) = e_v(t)-s_v(t)$

#### Question 2 : Donner dans le domaine temporel une relation liant$s_v(t)$, $\varepsilon_v(t)$ et leurs dérivées.

#### Réponse 2
Dans le domaine de la Laplace nous avons par la formule de Black : $\dfrac{S_v(p)}{E_v(p)}=H(p)$ Donc en temporel : $s_v(t)+T_v\dfrac{ds_v}{dt}=K_v\varepsilon_v$


### Discrétisation du signal
On souhaite effectuer une simulation numérique. Le pas de temps est noté $h$. On choisit $h=10^{-3}\; \text{s}$.
Dans le domaine numérique, on appelle $f_k$ le k-ième échantillon de la fonction $f$ à laquelle on associe la valeur $f(t_k)$ à l'instant $t_k=k \times h$.
On note $f_0=f(0)$. On rappelle que toutes les conditions initiales sont nulles.


#### Question 3
Montrer que dans le domaine numérique :
 - $\varepsilon_{vk}=e_{vk}-s_{vk}$;
 - $s_{vk}=\dfrac{1}{h+T_v}(T_v s_{vk-1} + K_v h \varepsilon_{vk})$.


#### Réponse 3
- Relation obtenue lors de la question 2, on a juste à rajouter des petits k : $\varepsilon_{vk}=e_{vk}-s_{vk}$
- On peut approximer la dérivée de $s_v(t)$ par $\dfrac{s_{vk-1}+s_{vk}}{h}$. La relation de la question 2 nous donne donc : $s_{vk}(t)+T_v\dfrac{s_{vk-1}+s_{vk}}{h}=K_v\varepsilon_{vk}$. D'où la relation demandée en isolant $s_{vk}$ : $s_{vk}=\dfrac{1}{h+T_v}(T_v s_{vk-1} + K_v h \varepsilon_{vk})$

Pour la question d'après on a besoin de changer la formule en remplaçant epsilon qui nous donne : $s_{vk}=\dfrac{T_vs_{vk-1}+K_vhe_{vk}}{h+T_v+K_vh}$

#### Question 4 
Implémenter un code Python qui effectue une simulation de l'asservissement en vitesse pour une entrée échelon d'amplitude 2 rad/s. Tracer la réponse simulée $s_v(t)$ et l'écart $\varepsilon_v(t)$ en fonction du temps $t$ sur un intervalle de 2 secondes.

```python
#Réponse 4
E0 = 2    # rad/s
svk = [0]
epsvk = [E0]
for k in range(1, int(2//h)):
    svk.append((Tv*svk[-1] + Kv*h*E0)/(h+Tv+Kv*h))
    epsvk.append(E0 - svk[-1])
plt.plot(epsvk, label=r"Ecart $\varepsilon(t)$")
plt.plot(svk, label=r"Sortie $s_v(t)$")
plt.legend()
plt.xlabel(r"Temps en $ms$")
plt.ylabel("Sortie en rad/s")
plt.show()
```

#### Question 11
Implémenter un code Python qui effectue une simulation de l'asservissement en vitesse avec correction PI pour une entrée échelon d'amplitude 2 rad/s.

```python
#Réponse 11
les_t   = [0]
les_sv  = [0]
les_eps = [0]
les_ev  = [2]
les_cv  = [0]
t=h
while t<2:
    les_t.append(t)
    les_eps.append(les_ev[-1]-les_sv[-1])
    les_cv.append(les_cv[-1]+Kp*(les_eps[-1]*(1+h/Ti)-les_eps[-2]))
    les_sv.append((1/(h+Tv))*(h*Kv*les_cv[-1]+Tv*les_sv[-1]))
    t=t+h
```

#### Question 12
Tracer la réponse simulée $s_v(t)$ en fonction du temps $t$.

```python
## Réponse 12
plt.plot(les_t, les_sv)
plt.axhline(1.90, color='black', linestyle='--')
plt.xlabel(r"Temps en $s$")
plt.ylabel(r"$s_v$ en $s$")
plt.axis([0,2,0,2])
plt.grid()
plt.show()
```
